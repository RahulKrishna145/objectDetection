#### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import joblib
```

#### Basic EDA

```python
df = pd.read_csv("data\\Human_poses.csv")
df.columns = df.columns.str.lstrip()
df.columns = df.columns.str.strip().str.replace(' ', '_')
```

```python
df.head()
```

```python
df.columns
```

#### Preprocessing

```python
df.drop(['c0_(Nose)','c1_(Neck)','c2_(RShoulder)','c3_(RElbow)','c4_(RWrist)','c5_(LShoulder)','c6_(LElbow)','c7_(LWrist)','c8_(RHip)','c9_(RKnee)','c10_(RFoot)','c11_(LHip)','c12_(LKnee)','c13_(LFoot)','c14_(REye)','c15_(LEye)','c16_(REar)','c17_(LEar)','class_(number)'], axis=1, inplace=True)
```

```python
df.head()
```

```python
le = LabelEncoder()
df['class_(text)']=le.fit_transform(df['class_(text)'])
df.head()
```

#### Creating Subsets

```python
X = df.drop(['Filename','class_(text)'],axis=1)
```

```python
X.head()
```

```python
Y=df[['class_(text)']]
```

```python
Y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
print("Input train size :",x_train.shape[0])
print("Input test size :",x_test.shape[0])
print("Output train size :",y_train.shape[0])
print("Output test size :",y_test.shape[0])
```

#### Training classifier and evaluating results

```python
cla = RandomForestClassifier(n_estimators=100, random_state=42)
cla.fit(x_train,y_train)
```

```python
y_pred = cla.predict(x_test)
```

```python
score = accuracy_score(y_test, y_pred)
print("accuracy is :",score)
```

#### exporting trained classifier and column names

```python
%%capture
joblib.dump(cla, "pkl_files\\rf_pretrained.pkl")
joblib.dump(x_train.columns.to_list(), "pkl_files\\column_names.pkl")
```
